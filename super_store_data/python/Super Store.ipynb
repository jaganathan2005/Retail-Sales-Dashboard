```python
import pandas as pd

# Step 1 - Load file
df = pd.read_csv(r"C:\\Users\\JAGAN\\Documents\\projects\\Sample - Superstore.csv", encoding='latin1')

# Step 2 - Drop garbage columns
df = df.drop(columns=['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23'])
print("✅ Garbage columns dropped!")

# Step 3 - Fix date formats
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='mixed', dayfirst=True)
print("✅ Dates fixed!")

# Step 4 - Days to Ship FIRST (before removing time!)
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days
print("✅ Days to Ship added!")
print(df['Days to Ship'].head(10))

# Step 5 - NOW remove time part
df['Order Date'] = df['Order Date'].dt.date
df['Ship Date'] = df['Ship Date'].dt.date
print("✅ Time part removed!")

# Step 6 - Profit Margin %
df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)
print("✅ Profit Margin % added!")

# Step 7 - Missing values check
print("\n✅ Missing Values Check:")
print(df.isnull().sum())

# Step 8 - Save clean file
df.to_excel(r"C:\Users\Jagan\Downloads\Superstore_Cleaned1.xlsx", index=False)
print("\n🎉 Clean file saved!")
```
